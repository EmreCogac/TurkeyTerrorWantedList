```python
import json
import os
import requests
import re
```

```python

# JSON dosyasının yolunu belirtiyoruz.
json_file_path = "data.json"

# Görsellerin kaydedileceği klasörü oluştur
output_folder = "teror"
os.makedirs(output_folder, exist_ok=True)

# JSON dosyasını aç ve içeriğini oku
with open(json_file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

# Ana URL'yi tanımla
base_url = "https://www.terorarananlar.pol.tr"
```

```python
for entry in data["kirmizi"]:
    if "GorselURL" in entry and entry["GorselURL"]:
        clean_orgut_adi = re.sub(r'[<>:"/\\|?*]', '-', entry["TOrgutAdi"])
        if entry.get("DogumYeri"):
            clean_dogum_yeri = re.sub(r'[<>:"/\\|?*]', '-', entry["DogumYeri"])
        else:
            clean_dogum_yeri = "Unknown"  
        custom_name = f"{entry['Adi']}-{entry['Soyadi']}-{clean_dogum_yeri}-{clean_orgut_adi}-{entry['TKategoriAdi']}"

        for index, image_path in enumerate(entry["GorselURL"]):
            full_url = base_url + image_path
            try:
                response = requests.get(full_url, stream=True)
                response.raise_for_status()  
                image_extension = os.path.splitext(image_path)[1]
                image_name = f"{custom_name}{f'_{index}' if index > 0 else ''}{image_extension}"
                output_folder = f'teror/{entry["TKategoriAdi"]}/{clean_orgut_adi}/{entry["Adi"]}-{entry["Soyadi"]}'
                os.makedirs(output_folder, exist_ok=True)
                file_path = os.path.join(output_folder, image_name)
                if os.path.exists(file_path):
                    print(f"{file_path} zaten mevcut, atlanıyor.")
                    continue 
                with open(file_path, "wb") as image_file:
                    for chunk in response.iter_content(1024):
                        image_file.write(chunk)

                print(f"{image_name} başarıyla indirildi.")
            except requests.exceptions.RequestException as e:
                print(f"{custom_name} indirilemedi. Hata: {e}")
            except FileNotFoundError:
                print(f"{file_path} için dizin bulunamadı. Lütfen dizin yapısını kontrol edin.")
            except OSError as e:
                print(f"{image_name} kaydedilemedi. OSError: {e}")
            except Exception as e:
                print(f"{image_name} kaydedilemedi. Hata: {e}")
    else:
        print(f"{entry['Adi']} {entry['Soyadi']} kaydında fotoğraf bulunamadı, atlanıyor.")
```
